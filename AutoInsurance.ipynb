```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

insurance = pd.read_csv("AutoInsurance.csv",sep=",",header=0)
#insurance.info() # 9134 entries
insurance.head(1000)

#print(insurance['Customer'].unique().size) #9134, each customer is unique
#print(insurance['State'].unique()) # data is from 5 states 'Washington' 'Arizona' 'Nevada' 'California' 'Oregon'
insurance['Customer Lifetime Value']=insurance['Customer Lifetime Value'].round(2)
#print(insurance['Response'].unique()) - Yes, No
#print(insurance['Coverage'].unique()) - Basic, Extended, Premium
#print(insurance['Education'].unique()) - 'Bachelor' 'College' 'Master' 'High School or Below' 'Doctor'

#print(insurance['Effective To Date'].unique()) - Format varies, some data is mm/dd/yy and some are mm/dd/yyyy
insurance['Effective To Date'] = pd.to_datetime(insurance['Effective To Date'])

#print(insurance['EmploymentStatus'].unique()) -- ['Employed' 'Unemployed' 'Medical Leave' 'Disabled' 'Retired']
#print(insurance['Gender'].unique()) -- 'F' 'M'

#print(insurance['Income'].unique())
#print(insurance['Location Code'].unique())
#print(insurance['Marital Status'].unique())
#print(insurance['Monthly Premium Auto'].unique())
#print(insurance['Months Since Last Claim'].unique())

#print(insurance['Months Since Policy Inception'].unique())
#print(insurance['Number of Open Complaints'].unique())
#print(insurance['Number of Policies'].unique())
#print(insurance['Policy Type'].unique()) --['Corporate Auto' 'Personal Auto' 'Special Auto']
#print(insurance['Policy'].unique()) -- ['Corporate L3' 'Personal L3' 'Corporate L2' 'Personal L1' 'Special L2''Corporate L1' 'Personal L2' 'Special L1' 'Special L3']
#print(insurance['Renew Offer Type'].unique())
#print(insurance['Sales Channel'].unique())
#print(insurance['Total Claim Amount'].unique()) -- round off to two decimal placess
insurance['Total Claim Amount'] = insurance['Total Claim Amount'].round(2)

#print(insurance['Vehicle Class'].unique())
#print(insurance['Vehicle Size'].unique())
```

```python
#Visualization
insurance.columns
# ['Customer', 'State', 'Customer Lifetime Value', 'Response', 'Coverage','Education', 'Effective To Date', 
#'EmploymentStatus', 'Gender','Income', 'Location Code', 'Marital Status', 'Monthly Premium Auto',
#'Months Since Last Claim', 'Months Since Policy Inception','Number of Open Complaints', 'Number of Policies', 
#'Policy Type','Policy', 'Renew Offer Type', 'Sales Channel', 'Total Claim Amount','Vehicle Class', 'Vehicle Size']

# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10, 6))
corr = insurance.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm")
f.subplots_adjust(top=0.93)
t= f.suptitle('Insurance Attributes Correlation Heatmap', fontsize=14)
#Monthly premium is highly correlated to the Total Claim Amount
```

```python
state=insurance.groupby(['State'])['Total Claim Amount'].mean()
state.plot(figsize=(20,10),title="Average Claim by State")
```

```python
stateTotal=insurance.groupby(['State'])['Total Claim Amount'].sum()
stateTotal.plot(figsize=(20,10),title="Total Claim by State")
#### Nevada has the highest average claim, whereas the highest total claim amount was in California
#### How does the state of the insurer affect the claim ???
```
